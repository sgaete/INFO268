## 4. Escribir datos de Twitter en una Base de Datos Mongo

```python
from pymongo import MongoClient
client = MongoClient()
db = client.INFO268
results = twitter.search(q="valdivia")
for tweet in results['statuses']:
    baked = {"fecha":tweet["created_at"], \
           "tweet": tweet['text'], \
           "username": tweet['user']['name'], \
           "location": tweet['user']['location']}
    #print(baked)
    db.tweets.insert_one(baked)
    
    
```

```python
from pymongo import MongoClient
import pprint
client = MongoClient()
db = client.INFO268
print ("Parte I, 1)Mostrar los documentos de su colección mostrando solamente el campo \"location\"." )
cursor = db.tweets.find()
for t in cursor:
    if (t['location']!=''):
        print (t['location'])
    else:
        print('None')    
```

```python
print ("Parte I, 2)Mostrar los tweets que contienen la palabra \"UACh\".")
cursor = db.tweets.find({"$or":[{"tweet":{"$regex":"uach", "$options": "i"}},\
                                {"location":{"$regex":"uach", "$options": "i"}} ]})
for t in cursor:
    pprint.pprint(t)
```

```python
print ("Parte I, 3)Contar el número de tweets por username.")
cursor = db.tweets.aggregate([ {"$group" : {"_id":"$username", "count":{"$sum":1} } } ])
for t in cursor:
    pprint.pprint(t)
```
