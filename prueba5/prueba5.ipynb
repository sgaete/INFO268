```python
import pandas as pd
import matplotlib.pyplot as plt
dic=pd.read_csv('dictionary.csv',sep=',')
win=pd.read_csv('winter.csv',sep=',')
summ=pd.read_csv('summer.csv',sep=',')
```

<p><b>1. </b>En un histograma, visualizar el GDP per Capita de cada pa√≠s. Bonus: ordenar el histograma en orden decreciente.</p>

```python
histo=dic[['Country','GDP per Capita']].sort_values(by=['GDP per Capita'],ascending=False)
histo[:50].plot(kind='bar',figsize=(25,10),x='Country',fontsize=20)
```

¬øCu√°l es el n√∫mero de medalla por cada pa√≠s, sin diferenciar el tipo de medalla, en los juegos de verano? Visualizar en un histograma.

```python
medal=summ['Country'].value_counts()
medal[:50].plot(kind='bar',figsize=(30,7),fontsize=20)
```

¬øCu√°l es el n√∫mero de medalla por cada pa√≠s, diferenciando el color de la medalla, en los juegos de verano? Visualizar en un histograma.

```python
sumb=summ.groupby('Medal').get_group('Bronze')['Country'].value_counts().reset_index()
sums=summ.groupby('Medal').get_group('Silver')['Country'].value_counts().reset_index()
sumg=summ.groupby('Medal').get_group('Gold')['Country'].value_counts().reset_index()

m1 = pd.merge(sumb,sums,on='index')
m2 = pd.merge(m1, sumg, on='index')
m3 = m2.rename(index=str, columns={"Country_x": "Bronze","Country_y":"Silver","Country":"Gold"})
m3[:50].plot(kind='bar',stacked=True,figsize=(30,7),fontsize=20,x='index')
```

```python
winb=win.groupby('Medal').get_group('Bronze')['Country'].value_counts().reset_index()
wins=win.groupby('Medal').get_group('Silver')['Country'].value_counts().reset_index()
wing=win.groupby('Medal').get_group('Gold')['Country'].value_counts().reset_index()
m4=pd.merge(m2,winb,on='index')
m5=pd.merge(m4,wins,on='index')
m6=pd.merge(m5,wing,on='index')
m6
```
